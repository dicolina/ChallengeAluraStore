### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tienda['Total'] = tienda['Precio'] + tienda['Costo de envío']
facturacion_total1 = tienda['Total'].sum()

print(f"Facturación Total: ${facturacion_total1:.2f}")
```

```python
tienda2['Total'] = tienda2['Precio'] + tienda2['Costo de envío']
facturacion_total2 = tienda2['Total'].sum()

print(f"Facturación Total: ${facturacion_total2:.2f}")
```

```python
tienda3['Total'] = tienda3['Precio'] + tienda3['Costo de envío']
facturacion_total3 = tienda3['Total'].sum()

print(f"Facturación Total: ${facturacion_total3:.2f}")
```

```python
tienda4['Total'] = tienda4['Precio'] + tienda4['Costo de envío']
facturacion_total4 = tienda4['Total'].sum()

print(f"Facturación Total: ${facturacion_total4:.2f}")
```

```python
facturaciones = [facturacion_total1, facturacion_total2, facturacion_total3, facturacion_total4]
facturaciones
```

```python
categorias = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ventas = [facturacion_total1, facturacion_total2, facturacion_total3, facturacion_total4]
colores = ['blue', 'green', 'red', 'purple']

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(8, 5))

# Crear barras
barras = ax.bar(categorias, ventas, color=colores)

# Añadir título y etiquetas
ax.set_title('Facturacion por tienda', fontsize=14)
ax.set_xlabel('Tiendas', fontsize=12)
ax.set_ylabel('Facturacion', fontsize=12)

# Añadir valores encima de las barras
for barra in barras:
    height = barra.get_height()
    ax.text(barra.get_x() + barra.get_width()/2., height,
            f'{height}',
            ha='center', va='bottom')

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

Analisis de Ventas por categoria de la Tienda numero 1


```python
# Agrupar por categoría y calcular el total de ingresos
ventas_por_categoria = tienda.groupby('Categoría del Producto')['Total'].sum().sort_values(ascending=False)

print("Ventas por Categoría:")
print(ventas_por_categoria)

# Gráfico de Torta para Ventas por Categoría
plt.figure(figsize=(10, 8))
plt.pie(ventas_por_categoria, labels=ventas_por_categoria.index, autopct='%1.1f%%', startangle=90)
plt.title('Ventas por Categoría')
plt.show()
```

# 4. Productos más y menos vendidos
